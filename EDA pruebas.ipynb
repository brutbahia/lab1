```python
import pandas as pd
import ast


df = pd.read_csv('movies_dataset_modificado.csv', low_memory=False)
```

```python
type(df.at[1,"name_Genres"])
```

```python
# Convertir las cadenas de texto en listas
df['name_Genres'] = df['name_Genres'].apply(lambda x: ast.literal_eval(x))
```

```python
# Convertir la lista de géneros en filas
df_explode = df.explode('name_Genres')

# Contar cada género y hacer un gráfico de barras
plt.figure(figsize=(20, 8))
counts = df_explode['name_Genres'].value_counts()
plt.bar(counts.index, counts.values)
plt.xlabel('Género')
plt.ylabel('Cantidad')
plt.title('Cantidad de películas por género')
plt.show()
```

```python
from sklearn.metrics.pairwise import pairwise_distances
print(df.at[7,"id_Genres"])
```

```python
#Uno de los intentos de perfeccionar la funcion recomendacion
def get_movie_recommendations(titulo:str):

    df1 = df.sample(n=5000)
    # Obtenga el índice de la película en la similarity matrix 
    
    movie_similarity =  pairwise_distances(df1.pivot_table(index='title',columns=['id_Genres']).fillna(0), metric='cosine')
    movie_index = df[df['title'] == titulo].index[0]

    # Get the similarity scores for all movies compared to the given movie
    similarity_scores = movie_similarity[movie_index]

    # Sort the scores from highest to lowest
    similar_movie_indices = np.argsort(-similarity_scores)

    # Obtén las mejores películas más similares
    top_movie_indices = similar_movie_indices[1:6]

    # Obtén los títulos de las mejores películas más similares
    respuesta = [df.iloc[index]['title'] for index in top_movie_indices]
    

    return {'lista recomendada': respuesta }
```

```python
get_movie_recommendations("Grumpier Old Men")
```
