```python
import pandas as pd
import json
import ast
import numpy as np

# Carga el archivo CSV en un DataFrame de pandas
df = pd.read_csv("E:\hENRYDATASEMANAS\lab1\movies_dataset.csv")
```

```python
#Desanidar la columna "belongs_to_collection"
#Rellenamos los Nulos
df["belongs_to_collection"].fillna( "{'id': None, 'name': None, 'poster_path': None, 'backdrop_path': None}", inplace=True)
#Movemos a un data frame para poder separarlo
df1= pd.DataFrame(df['belongs_to_collection'])

# Función para dividir la cadena en columnas
def split_collection(collection_str):
    try:
        # Función para dividir la cadena en columnas
        collection_dict = ast.literal_eval(collection_str)
        # Extraer los valores correspondientes utilizando el método get()
        id = collection_dict.get('id')
        name = collection_dict.get('name')
        poster_path = collection_dict.get('poster_path')
        backdrop_path = collection_dict.get('backdrop_path')
        return pd.Series([id, name, poster_path, backdrop_path])
    except:
        return pd.Series([None, None, None, None])

# Aplicar la función a la columna y agregar las nuevas columnas al DataFrame
df[['id_collection', 'name_collection', 'poster_path_collection', 'backdrop_path_collection']] = df['belongs_to_collection'].apply(split_collection)
# Eliminar la columna original
df.drop('belongs_to_collection', axis=1, inplace=True)
df.head()
```

```python
#Desanidar la columna "'genres'"
# Definir función para separar lista de diccionarios en dos columnas
def split_genres(row):
    # Convertir string a lista de diccionarios
    genres_list = ast.literal_eval(row)
    # Crear listas separadas para id y name
    id_list = [d.get('id') for d in genres_list]
    name_list = [d.get('name') for d in genres_list]
    # Crear diccionario de salida
    out_dict = {'id_Genres': id_list, 'name_Genres': name_list}
    return pd.Series(out_dict)

# Aplicar función a cada fila del DataFrame y agregar columnas resultantes
df[['id_Genres', 'name_Genres']] = df['genres'].apply(split_genres)
# Eliminar la columna original
df.drop('genres', axis=1, inplace=True)
df.head()
```

```python
#Desanidar la columna "production_companies"
#Rellenamos los Nulos
df["production_companies"].fillna( "[{'name': 'sin dato', 'id': 'sin dato'}]", inplace=True)
# Definir función para separar lista de diccionarios en dos columnas
def split_production_companies(row):
   try:
        # Convertir string a lista de diccionarios
        production_companies_list = ast.literal_eval(row)
        # Crear listas separadas para id y name
        name_list = [d.get('name') for d in production_companies_list]
        id_list = [d.get('id') for d in production_companies_list]    
        # Crear diccionario de salida
        out_dict = {'name_production_companies': name_list, 'id_production_companies': id_list}
        return pd.Series(out_dict)
   except:
        salida = {'name_production_companies': None , 'id_production_companies': None}
        return pd.Series(salida)

# aplicar la función a la columna y agregar las nuevas columnas al DataFrame
df[['name_production_companies', 'id_production_companies']] = df['production_companies'].apply(split_production_companies)
# Eliminar la columna original
df.drop("production_companies", axis=1, inplace=True)
df.head()
```

```python
#Desanidar la columna "production_companies"
#Rellenamos los Nulos
df["production_countries"].fillna( "[{ 'iso_3166_1': None,'name': None}]", inplace=True)
# Definir función para separar lista de diccionarios en dos columnas
def split_production_countries(row):
   try:
        # Convertir string a lista de diccionarios
        production_countries_list = ast.literal_eval(row)
        # Crear listas separadas para id y name
        iso_3166_1_list = [d.get('iso_3166_1') for d in production_countries_list] 
        name_list = [d.get('name') for d in production_countries_list]
           
        # Crear diccionario de salida
        out_dict = {'name_production_countries': name_list, 'iso_3166_1_production_countries': iso_3166_1_list}
        return pd.Series(out_dict)
   except:
        salida = {'name_production_countries': None , 'iso_3166_1_production_countries': None}
        return pd.Series(salida)

# aplicar la función a la columna y agregar las nuevas columnas al DataFrame
df[['name_production_countries', 'iso_3166_1_production_countries']] = df['production_countries'].apply(split_production_countries)
# Eliminar la columna original
df.drop("production_countries", axis=1, inplace=True)
df.head()
```

```python
#Desanidar la columna spoken_languages
#Rellenamos los Nulos
df["spoken_languages"].fillna( "[{ 'iso_639_1': None,'name': None}]", inplace=True)
# Definir función para separar lista de diccionarios en dos columnas
def split_spoken_languages(row):
   try:
        # Convertir string a lista de diccionarios
        spoken_languages_list = ast.literal_eval(row)
        # Crear listas separadas para id y name
        iso_639_1_list = [d.get('iso_639_1') for d in spoken_languages_list] 
        name_list = [d.get('name') for d in spoken_languages_list]
           
        # Crear diccionario de salida
        out_dict = {'name_spoken_languages': name_list, 'iso_639_1_spoken_languages': iso_639_1_list}
        return pd.Series(out_dict)
   except:
        salida = {'name_spoken_languages': None , 'iso_639_1_spoken_languages': None}
        return pd.Series(salida)

# aplicar la función a la columna y agregar las nuevas columnas al DataFrame
df[['name_spoken_languages', 'iso_639_1_spoken_languages']] = df['spoken_languages'].apply(split_spoken_languages)
# Eliminar la columna original
df.drop("spoken_languages", axis=1, inplace=True)
df.head()
```

```python
#Verificar nulos
# contar los valores nulos en 'revenue' y 'budget'
null_counts1 = df['revenue'].isnull().sum()
null_counts2 = df['budget'].isnull().sum()
# imprimir el resultado
print('nulos en revenue', null_counts1 ,'nulos en budget',null_counts2)
#Remplazar los nulos por  0
df['revenue'].fillna(0, inplace=True)
df['budget'].fillna(0, inplace=True)
# contar los valores nulos en 'revenue' y 'budget'
null_counts1 = df['revenue'].isnull().sum()
null_counts2 = df['budget'].isnull().sum()
# imprimir el resultado
print('nulos en revenue', null_counts1 ,'nulos en budget',null_counts2)
```

```python
# Convertir columna release_date al formato deseado si esta fuera de formato de NaN
df["release_date"] = pd.to_datetime(df["release_date"], format="%Y-%m-%d", errors="coerce")

# Extraer el año de la fecha y crear la columna release_year
df["release_year"] = df["release_date"].dt.year
df["release_month"] = df["release_date"].dt.strftime("%B")
df['release _day'] = df['release_date'].dt.strftime('%A')
meses = {
    "January": "Enero",
    "February": "Febrero",
    "March": "Marzo",
    "April": "Abril",
    "May": "Mayo",
    "June": "Junio",
    "July": "Julio",
    "August": "Agosto",
    "September": "Septiembre",
    "October": "Octubre",
    "November": "Noviembre",
    "December": "Diciembre"
}

# Diccionario con los nombres de los días de la semana en inglés y español
dias = {
    "Monday": "Lunes",
    "Tuesday": "Martes",
    "Wednesday": "Miércoles",
    "Thursday": "Jueves",
    "Friday": "Viernes",
    "Saturday": "Sábado",
    "Sunday": "Domingo"
}

# Aplicar el método map() para cambiar los nombres en inglés por los nombres en español
df["release_month"] = df["release_month"].map(meses)
df["release _day"] = df["release _day"].map(dias)
# contar los valores nulos en 'release_date'
null_counts_year = df['release_date'].isnull().sum()
# imprimir el resultado
print(null_counts_year)
```
